```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
file_path = "/home/jovyan/research/Current/SNcount.csv"
df = pd.read_csv(file_path)

# Sort data for cumulative distribution
df_sorted_g = df.sort_values("G count")
df_sorted_r = df.sort_values("R count")

# Compute cumulative distribution
cdf_g = np.cumsum(df_sorted_g["G count"]) / np.sum(df_sorted_g["G count"])
cdf_r = np.cumsum(df_sorted_r["R count"]) / np.sum(df_sorted_r["R count"])

# Create cumulative distribution plots
plt.figure(figsize=(10, 5))
plt.plot(df_sorted_g["G count"], cdf_g, label="Green Band", color="green")
plt.xlabel("Number of Detections (G band)")
plt.ylabel("Cumulative Fraction")
plt.title("Cumulative Distribution of Detections - Green Band")
plt.legend()
plt.grid(True)
plt.xlim(0,50)
plt.show()


plt.figure(figsize=(10, 5))
plt.plot(df_sorted_r["R count"], cdf_r, label="Red Band", color="red")
plt.xlabel("Number of Detections (R band)", fontsize=14)
plt.ylabel("Cumulative Fraction", fontsize=14)
plt.title("Cumulative Distribution of Detections - Red Band", fontsize=15)
plt.legend()
plt.grid(True)
plt.xlim(0, 5)
plt.tick_params(axis='both', labelsize=14)  # << add this
#plt.savefig('cumulativeR.png')
plt.show()
```
